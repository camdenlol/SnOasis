```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
url = 'https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(url, parse_dates=['Date'])

# Extract the 'final price' column
final_price_column = df['Final price (USD)']

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Plot the 'final price' as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(df.index, final_price_column, color='b', width=20)  # Adjust width as needed

# Set labels and title
plt.title('Final Price by Date')
plt.xlabel('Date')
plt.ylabel('Final Price')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
url = 'https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(url, parse_dates=['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Resample the data to monthly increments and sum the 'final price'
monthly_final_sales = df['Final price (USD)'].resample('M').sum()

# Plot the total final sales as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(monthly_final_sales.index, monthly_final_sales, color='b', width=20)  # Adjust width as needed

# Set labels and title
plt.title('Total Final Sales per Month')
plt.xlabel('Month')
plt.ylabel('Total Final Sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
url = 'https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(url, parse_dates=['Date'])

# Extract day of the week from the 'Date' column
df['Day_of_Week'] = df['Date'].dt.day_name()

# Group by day of the week and sum the 'Quantity'
daily_quantity = df.groupby('Day_of_Week')['Quantity'].sum()

# Sort days of the week in the correct order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_quantity = daily_quantity.reindex(days_order)

# Plot the quantity sold per day of the week as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(daily_quantity.index, daily_quantity, color='purple')

# Set labels and title
plt.title('Total Quantity Sold per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity Sold')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
url = 'https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(url, parse_dates=['Date'])

# Extract day of the week from the 'Date' column
df['Day_of_Week'] = df['Date'].dt.day_name()

# Group by day of the week and sum the 'final price'
daily_final_price = df.groupby('Day_of_Week')['Final price (USD)'].sum()

# Sort days of the week in the correct order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_final_price = daily_final_price.reindex(days_order)

# Plot the final price per day of the week as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(daily_final_price.index, daily_final_price, color='orange')

# Set labels and title
plt.title('Total Final Price per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Final Price')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
